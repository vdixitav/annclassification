```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
```

```python
# load dataset
data=pd.read_csv("Churn_Modelling.csv")
print(data)
```

```python
# pre-processing
# drpoong irrelatvant coloumn

data=data.drop(['RowNumber','CustomerId','Surname'],axis=1)
data
```

```python
data.info()
```

```python
# eNCODING OBJECTIVES TO NUBERS

label_encoder_gender=LabelEncoder()
data['Gender']=label_encoder_gender.fit_transform(data['Gender'])
data
```

```python
data.value_counts()
```

```python
data.describe
```

```python
from sklearn.preprocessing import OneHotEncoder

# Initialize the OneHotEncoder
onehot_encoder_geo = OneHotEncoder()

# Reshape the Geography column to a 2D array
geography_reshaped = data['Geography'].values.reshape(-1, 1)

# Fit and transform the reshaped Geography column
geo_encoded = onehot_encoder_geo.fit_transform(geography_reshaped)

# Convert the result to an array if needed
geo_encoded_array = geo_encoded.toarray()

# Display the encoded result
geo_encoded_array
```

```python
onehot_encoder_geo.get_feature_names_out(['Geography'])
```

```python
geo_encoded_df=pd.DataFrame(geo_encoded.toarray(),columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
# combining all one hoted  column with original data
data=pd.concat([data.drop('Geography',axis=1),geo_encoded_df],axis=1)
```

```python
data.head()
```

```python
# save the encoder and scaler

with open('label_encoder_gender.pkl','wb') as file:
    pickle.dump(label_encoder_gender,file)


with open('onehot_encoder_geo.pkl','wb') as file:
      pickle.dump(onehot_encoder_geo,file)  
```

```python
# divide the dataset into independed and depended featre

data.head()
```

```python
X=data.drop('Exited',axis=1)
Y=data['Exited']
print(X)
```

```python
print(Y)
```

```python
# split data into training and testing

x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
with open('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

ANN Implementaion


```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping ,TensorBoard
import datetime

```

```python
# build model

model=Sequential([
    Dense(64,activation='relu',input_shape=(x_train.shape[1],)),## 1st hidden layer connected with i/p
    Dense(32,activation='relu'),#HL 2
    Dense(1,activation='sigmoid')
])
```

```python
model.summary()
```

```python
import tensorflow as tf

opt = tf.keras.optimizers.Adam(learning_rate=0.01)
loss=tf.keras.losses.BinaryCrossentropy()
loss
```

```python
# Compile the model
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=['accuracy'])
```

```python
# set up the tensorboard
log_dir="lofs/fit"+datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callback=TensorBoard(log_dir,histogram_freq=1)
```

```python
# setup Early Stopping 
early_stopping_callback=EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
```

```python
# Train the model
# Train the model
history = model.fit(
    x_train, y_train, 
    validation_data=(x_test, y_test), 
    epochs=100,
    callbacks=[tensorflow_callback, early_stopping_callback])
```

```python
model.save('model.h5')
```
